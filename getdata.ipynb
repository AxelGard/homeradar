```python

html_content = ""

with open("./hemnet.html") as f: 
    html_content = f.read()
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all spans with the specific class name
spans = spans = soup.select('span[class*=ForSaleAttributes_primaryAttributes]')
# Extract the text content of each span
values = [span.get_text() for span in spans]
values = [v.replace(" ", "") for v in values]
values = [v.replace("m²", "") for v in values]
values = [v.replace("rum", "") for v in values]
values = [v.replace("vån", "") for v in values]
values = [v.replace("kr", "") for v in values] 
values = [v.replace(",", ".") for v in values] 
```

```python
i = 0
props = []
while i < len(values)-4: 
    props.append(
        {
            "price":values[i],
            "size": values[i+1],
            "rooms": values[i+2],
            "hight": values[i+3],
        }
    )
    i+= 4
props
```
