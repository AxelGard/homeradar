```python

html_content = ""

with open("./hemnet.html") as f: 
    html_content = f.read()
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

spans = soup.select('span[class*=ForSaleAttributes_primaryAttributes__tqSRJ]')

values = [span.get_text() for span in spans]
values = [v.replace(" ", "") for v in values]
values = [v.replace("m²", "") for v in values]
values = [v.replace("rum", "") for v in values]
values = [v.replace("vån", "") for v in values]
values = [v.replace("kr", "") for v in values] 
values = [v.replace(",", ".") for v in values] 
len(values)/4
```

```python
i = 0
props = []
while i < len(values)-4: 
    props.append(
        {
            "price":values[i],
            "size": values[i+1],
            "rooms": values[i+2],
            "hight": values[i+3],
        }
    )
    i+= 4
len(props), props[0]
```

```python
from geopy.geocoders import Nominatim
import time 

app = Nominatim(user_agent="tutorial")
address_cache = {}
def get_address_info(address:str)->dict:
    if address in address_cache:
        return address_cache[address]
    time.sleep(1)
    try:
        raw = app.geocode(address).raw
        address_cache[address] = raw 
        return raw
    except:
        return get_address_info(address)
```
